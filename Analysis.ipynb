```python
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils import shuffle
from sklearn.svm import SVC
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def dt_model_analysis(x):
    
    pkl_file = open( x, 'rb')
    
    mydict = pickle.load(pkl_file)
    
    pkl_file.close()
    
    df = pd.DataFrame() 
    
    df['mean_fit_time']=mydict['mean_fit_time']
    
    df['mean_score_time']=mydict['mean_score_time']
    
    df['mean_test_accuracy']=mydict['mean_test_accuracy']
    
    df['mean_test_f1']=mydict['mean_test_f1']
    
    df['mean_test_precision']=mydict['mean_test_precision']
    
    df['mean_test_recall']=mydict['mean_test_recall']
    
    df['mean_train_accuracy']=mydict['mean_train_accuracy']
    
    df['mean_train_f1']=mydict['mean_train_f1']
    
    df['mean_train_precision']=mydict['mean_train_precision']
    
    df['mean_train_recall']=mydict['mean_train_recall']
    
    df['max_depth']=mydict['param_max_depth'].astype(np.float64)
    
    df['max_features']=mydict['param_max_features']
    
    df['min_leaf_samples']=mydict['param_min_samples_leaf'].astype(np.float64)
    
    df['min_samples_split']=mydict['param_min_samples_split'].astype(np.float64)
    
    df['std_test_accuracy']=mydict['std_test_accuracy']
    
    df['std_test_f1']=mydict['std_test_f1']
    
    df['std_test_precision']=mydict['std_test_precision']
    
    df['std_test_recall']=mydict['std_test_recall']
    
    df=df[:440]
    
    print(sns.pairplot(df,hue='max_depth'))
    
    print(sns.lmplot(x='mean_fit_time',y='mean_test_recall',data=df,hue='max_depth',fit_reg=False))
    
    print(sns.lmplot(x='mean_score_time',y='mean_test_recall',data=df,hue='max_depth',fit_reg=False))
    
    print(sns.lmplot(x='std_test_recall',y='mean_test_recall',data=df,hue='max_depth',fit_reg=False))
    
    print(sns.heatmap(df.corr()))
    
    return df,mydict
```

## Loading Dataset

```python
# Loading the dataset
X = pd.read_csv('normalized_generic.csv',sep='\t')
X = X.drop('Unnamed: 0',axis=1)

a=X.loc[X['rumor'] == 0.0]
b=X.loc[X['rumor'] == 1.0]
a=a[0:len(b)]
X=pd.concat([a,b])
X = shuffle(X)
X= X.reset_index(drop=True)

# Defining the last column of the dataframe as the output column of the model
y=X['rumor']
X = X.iloc[:, :-1]
```

## SVM Feature Investigation

```python
svm=SVC(C=1,gamma=0.9,kernel='linear')
svm.fit(X,y)
```

```python
sorted(svm.coef_[0])
```

```python
sorted_columns=np.array([x for _, x in sorted(zip(svm.coef_[0], X.columns))])
sorted_coef=np.array(sorted(svm.coef_[0]))
selection = [0,1,2,3,51,52]
sorted_coef=sorted_coef[selection]
sorted_columns=sorted_columns[selection]
```

```python
def f_importances(coef, names):
    imp = coef
    imp,names = zip(*sorted(zip(imp,names)))
    plt.barh(range(len(names)), imp, align='center')
    plt.yticks(range(len(names)), names)
    plt.show()

features_names = [X.columns]
f_importances(sorted_coef, sorted_columns)
```

## Decision Tree Feature Investigation

```python
dtree = DecisionTreeClassifier(max_depth=10,max_features='auto',min_samples_leaf=2,min_samples_split=2)
dtree.fit(X,y)
```

```python
sorted_columns=np.array([x for _, x in sorted(zip(dtree.feature_importances_, X.columns))])
sorted_coef=np.array(dtree.feature_importances_)
sorted_coef=sorted_coef[-10:]
sorted_columns=sorted_columns[-10:]
```

```python
f_importances(sorted_coef, sorted_columns)
```
